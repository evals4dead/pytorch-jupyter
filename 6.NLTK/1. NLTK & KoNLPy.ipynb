```python
import nltk

nltk.download("gutenberg")
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
```

```python
nltk.corpus.gutenberg.fileids()
```

```python
text = nltk.corpus.gutenberg.raw("shakespeare-macbeth.txt")
```

```python
print(text[:1000])
```

```python
text = text.lower()
```

```python
from nltk.tokenize import word_tokenize
print(word_tokenize(text[:500]))
```

```python
from nltk.tokenize import RegexpTokenizer
tokens = RegexpTokenizer("[\w]+").tokenize(text[:500]) # 특문 제거, remove special charaters
print(tokens)
```

```python
from nltk.corpus import stopwords
stopping = set(stopwords.words('english'))
print([token for token in tokens if not token in stopping]) # the a 와 같은 관사 제거
```

```python
print('the' in stopping, 'tragedie' in stopping)
```

```python
from nltk.stem import PorterStemmer
[PorterStemmer().stem(token) for token in tokens] # token들을 모두 PorterStemmer화 시킨다
```

```python
from nltk.tag import pos_tag
print(pos_tag(tokens)) #token들을 모두 pos tagging 한다.
```

```python
def tokenizer(tagger, doc):
    return ["/".join(p) for p in tagger(doc)]
```

```python
tokenizer(pos_tag, tokens) # 배열화 시킨다
```

# wordcloud

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def DrawWordCloud(data):
    wordcloud = WordCloud(stopwords=STOPWORDS, background_color='white', width=800, height=400).generate(data)
    plt.figure(figsize=(15, 25))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
DrawWordCloud(text)
```

# Words Frequency

```python
from collections import Counter
counts = dict(Counter(RegexpTokenizer("[\w]+").tokenize(text)).most_common(20))
```

```python
counts.items()
```

```python
import numpy as np

labels, values = zip(*counts.items())

indSort = np.argsort(values)[::-1]

print(indSort)

labels = np.array(labels)[indSort]
values = np.array(values)[indSort]

indexes = np.arange(len(labels))

plt.figure(figsize=(15, 5))
plt.xlabel('Top 15 Words')
plt.ylabel('Frequency')
plt.title('Word Frequency')

plt.bar(indexes, values)
plt.xticks(indexes, labels)
plt.show()
```
